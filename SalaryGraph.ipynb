```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

dataset = pd.read_excel("nba.xlsx")

dataset['Salary']= dataset ['Salary'].astype('float')

plt.figure(figsize=(65,110))

plt.title("Player's Salaries")

plt.ylabel("Name")

plt.plot(dataset.Salary,dataset.Name)

                             
plt.show()
```
